```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Parte 1 Carga y Limpieza

```
# Leer el archivo CSV
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

# Imprimir primeras 5 filas y tipos de datos
print("Primeras 5 filas del dataset:")
display(df.head())
print("\nTipos de datos de cada columna:")
print(df.dtypes)

# Reemplazar valores nulos en 'horsepower' con la mediana
mediana_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
print(f"\nValores nulos en horsepower reemplazados por la mediana: {mediana_hp}")

#Crear nueva columna: relacion_peso_potencia
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
print("\nColumna 'relacion_peso_potencia' creada correctamente.")


print(df[['name', 'relacion_peso_potencia']].head())
```

#Parte2: Visualización de Datos (Matplotlib)

```

# 1. Histograma: Para ver cómo se distribuye la aceleración
plt.figure(figsize=(8, 5))
plt.hist(df['acceleration'], bins=15, color='green', edgecolor='white')
plt.title('Distribución de la Aceleración en los Vehículos')
plt.xlabel('Segundos de Aceleración (acceleration)')
plt.ylabel('Número de Vehículos')
plt.show()

# 2. Gráfico de Dispersión: Relación entre motor y potencia
plt.figure(figsize=(8, 5))
plt.scatter(df['displacement'], df['horsepower'], alpha=0.6, color='teal')
plt.title('Relación entre Desplazamiento del Motor y Caballos de Fuerza')
plt.xlabel('Desplazamiento (displacement)')
plt.ylabel('Caballos de Fuerza (horsepower)')
plt.grid(True, linestyle='--', alpha=0.5) # Esto le da un toque profesional
plt.show()

# 3. Gráfico de Barras: Rendimiento promedio por país
# Primero agrupamos por el país de origen y sacamos el promedio de 'mpg'
rendimiento_por_pais = df.groupby('origin')['mpg'].mean()

plt.figure(figsize=(8, 5))
rendimiento_por_pais.plot(kind='bar', color=['navy', 'darkred', 'darkgreen'])
plt.title('Rendimiento Promedio (MPG) según el País de Origen')
plt.xlabel('País de Origen ')
plt.ylabel('Millas por Galón (Promedio)')
plt.xticks(rotation=0) # Para que los nombres de los países no salgan chuecos
plt.show()
```
